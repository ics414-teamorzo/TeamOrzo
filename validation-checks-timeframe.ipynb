```python
# imports

# https://www.tutorialspoint.com/python/python_gui_programming.htm
from tkinter import * 

master = Tk()
master.title("Calendering System")
```

```python
#### GRID NUMBERING #####
titleR = 0
titleC= 0

sumR = titleR + 4

startR = sumR + 1

endR = startR + 2

adR = endR + 2

pR = adR + 2

recR = pR + 1

buttonR = recR + 2
```

```python
def timeFrame():
    if ( int(year.get()) > int(yearE.get()) ): #eg if start year 2025 > than end year 2024 error
        Label(master, text='End year cannot be before start year').grid(row=buttonR+5, column=titleC, columnspan=5)
    elif( year.get()==yearE.get()):
        #if end month occurs before start month
        for i in range(len(months)-1):
            if months[i] == monthE.get() and months[i]!=month.get():
                Label(master, text='End mont cannot be before start month in the same year').grid(row=buttonR+6, column=titleC, columnspan=5)
        if month.get()==monthE.get() :
            if int(day.get())>int(dayE.get()) :
                Label(master, text='End day cannot be befroe starte day').grid(row=buttonR+7, column=titleC, columnspan=5)
            elif day.get()==dayE.get(): 
                if( int(hour.get())>int(hourE.get()) ):
                    Label(master, text='End time(hour) cannot be before start time').grid(row=buttonR+8, column=titleC, columnspan=5)
                elif hour.get()==hourE.get():
                       if( int(minute.get())>=int(minuteE.get()) ):
                            Label(master, text='End time(minute) cannot be before or at start time').grid(row=buttonR+9, column=titleC, columnspan=5)
    return 
```

```python
save = Button(master, text="SAVE", command=timeFrame)
save.grid(row=buttonR, column=6)
```
