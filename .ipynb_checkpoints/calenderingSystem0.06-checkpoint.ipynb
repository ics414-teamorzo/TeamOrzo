```python
####  INPUT VALIDITY FUNCTIONS  ###
def required():
    extraFeedback['text'] = " "
    if len(eventName.get()) == 0:
        extraFeedback['text']="Please enter a file name including a .ics"
        return False 
    if ".ics" not in eventName.get():
        extraFeedback['text']="Please enter a file name including a .ics"
        return False
    if year.get()=='----' and month.get()=='---' and day.get()=='--':
        extraFeedback['text']="Please specify start date"
        return False
    if hour.get()=='--' and minute.get()=='--': 
        extraFeedback['text']="Please specify start time"
        return False    
    if yearE.get()=='----' and monthE.get()=='---' and dayE.get()=='--':
        extraFeedback['text']="Please specify end date"
        return False
    if hourE.get()=='--' and minuteE.get()=='--': 
        extraFeedback['text']="Please specify end time"
        return False
    return True

def endOfMonth():
    # DTSTART
    if month.get() == 'Febuary':
        if year.get()=='2020' or year.get()=='2024' or year.get()=='2028': #leap year
            if day.get()=='30' or day.get()=='31': # invalid date
                extraFeedback['text']="Febuary only has 29 days for leap years"    
                return False
        else: #not a leap year
            if day.get()=='29' or day.get()=='30' or day.get()=='31': # invalid date
                extraFeedback['text']="Febuary only has 28 days, the year specified is not a leap year"        
                return False
    elif month.get()=='April' or month.get()=='June' or month.get=='September'or month.get()=='November':
        if day.get()=='31':
            extraFeedback['text']="%s only has 30 days"%month.get()
            return False
    # DTEND    
    if monthE.get() == 'Febuary':
        if yearE.get()=='2020' or yearE.get()=='2024' or yearE.get()=='2028': #leap year
            if dayE.get()=='30' or dayE.get()=='31': # invalid date
                extraFeedback['text']="Febuary only has 29 days for leap years"    
                return False
        else: #not a leap year
            if dayE.get()=='29' or dayE.get()=='30' or dayE.get()=='31': # invalid date
                extraFeedback['text']="Febuary only has 28 days, the year specified is not a leap year"        
                return False
    elif monthE.get()=='April' or monthE.get()=='June' or monthE.get=='September'or monthE.get()=='November':
        if dayE.get()=='31':
            extraFeedback['text']="%s only has 30 days"%monthE.get()
            return False
    return True

def timeFrame():
    if ( int(year.get()) > int(yearE.get()) ): #eg if start year 2025 > than end year 2024 error
        extraFeedback['text']="End year cannot be before start year"
        return False
    elif( year.get()==yearE.get()):
        #if end month occurs before start month
        for i in range(len(months)-1):
            if months[i] == monthE.get() and months[i]!=month.get():
                extraFeedback['text']="End mont cannot be before start month in the same year"
                return False
        if month.get()==monthE.get() :
            if int(day.get())>int(dayE.get()) :
                extraFeedback['text']="End day cannot be befroe starte day"
                return False
            elif day.get()==dayE.get(): 
                if( int(hour.get())>int(hourE.get()) ):
                    extraFeedback['text']="End time(hour) cannot be before start time"
                    return False
                elif hour.get()==hourE.get():
                       if( int(minute.get())>=int(minuteE.get()) ):
                            extraFeedback['text']="End time(minute) cannot be before or at start time"
                            return False
    return True
```

```python
####  FUNCTIONS  ####

#validity check, ignore if empty and not requiered
def check():
    testFeedback['text'] = " "
    if required() == False:
        testFeedback['text'] = "Not all requiered fields are filled out"
        return False
    if endOfMonth() == False:
        testFeedback['text'] = "Invalid month date"
        return False
    if timeFrame() == False:
        testFeedback['text'] = "End time cannot be before start time"
        return False
    return True

#edit existing file based on feilds, ignore feilds left blank
def edit(eventExisting):
    testFeedback['text'] = "to be implamented: edit existing file"
    # eventExisting.close()
    return
    
# create new file based on feilds, ignore feilds left blank
def create(eventNew):
    for i in range(len(months)):
        if months[i] == month.get():
            if i<10:
                monthNum = "0%i"%i
            else:
                    monthNum = '%i'%i
        if months[i] == monthE.get():
            if i<10:
                monthNumE = "0%i"%i
            else:
                monthNumE = '%i'%i
    if pri.get() == "None":
        priorityNum = '0'
    if pri.get() == "High":
        priorityNum = '1'
    if pri.get() == "Medium":
        priorityNum = '5'
    if pri.get() == "Low":
        priorityNum = '9' 
        
    ####  READ RECURANCE RULE FEILDS AND PUT ALL INTO A STRING  ####
    intervalBoo = False
    freqBoo = False
    occBoo = False

    if interval.get() !=  "--":
        intervalStr = "INTERVAL=%s"%interval.get()
        intervalBoo = True
    else:
        intervalStr = ""
        
    if freq.get() != "----":
        freqBoo = True
        if freq.get() == "Year(s)":
            freqStr = "FREQ=YEARLY"
        elif freq.get() == "Month(s)":
            freqStr = "FREQ=MONTHLY"
        elif freq.get() == "Week(s)":
            freqStr = "FREQ=WEEKLY"
        elif freq.get() == "Day(s)":
            freqStr = "FREQ=DAILY"
    else:
        freqStr = ""
            
    if occurence.get() != "--" and occurence.get() != "On going":
        occBoo = True
        occStr = "COUNT=%s"%occurence.get()
    else:
        occStr = ""
           
    byDayStr = ["SU", "MO", "TU", "WE", "TH", "FR", "SA"]
    byDayBoo = [su.get(), mo.get(), tu.get(), we.get(), th.get(), fr.get(), sa.get() ]

    numOfDays = 0
    for i in range(len(byDayBoo)):
        if byDayBoo[i] == 1:
            numOfDays += 1
            if numOfDays == 1:
                dayStr = "BYDAY=%s"%byDayStr[i]
            if numOfDays > 1:
                 dayStr += ",%s"%byDayStr[i]
    if numOfDays == 0:
        dayStr = "BYDAY=MO" #if no day of the week selected automatically chose Monday

    reoccuranceRule = "%s;WKST=SU;%s;%s;%s"%(freqStr, occStr, intervalStr, dayStr) 
    
    eventNew.write("BEGIN:VCALENDAR\n")
    eventNew.write("VERSION:2.0\n")
    eventNew.write("BEGIN:VEVENT\n")
    eventNew.write("DTSTART:%s%s%sT%s%s00\n"%( year.get(), monthNum, day.get(), hour.get(), minute.get() ) )
    eventNew.write("DTEND:%s%s%sT%s%s00\n"%( yearE.get(), monthNumE, dayE.get(), hourE.get(), minuteE.get() ) )
    eventNew.write("CLASS:%s\n"%access.get())
    eventNew.write("PRIORITY:%s\n"%priorityNum)
    eventNew.write("DESCRIPTION:%s\n"%desc.get())
    eventNew.write("SUMMARY:%s\n"%description.get())
    eventNew.write("LOCATION:%s\n"%street.get())
    eventNew.write("GEO:21.315060;-157.812460\n") ####  TO BE FILLED OUT BASED ON LOCAITON https://www.latlong.net/
    eventNew.write("TZID:America/Hawaii\n") ####  TO BE FILLED OUT BASED ON LOCAITON
    eventNew.write("RRULE:%s\n"%reoccuranceRule)
    eventNew.write("END:VEVENT\n")
    eventNew.write("END:VCALENDAR\n")
    eventNew.close()
    return
    
# clear all feilds and inputs, so you can add/edit a new event
def clearFeilds():
    eventName.delete(0,'end')
    # clear DTSTART
    year.set('----')
    month.set('---')
    day.set('--')
    hour.set('--')
    minute.set('--')
    #clear DTEND
    yearE.set('----')
    monthE.set('---')
    dayE.set('--')
    hourE.set('--')
    minuteE.set('--')
    #clear summary and description
    desc.delete(0, 'end')
    description.delete(0, 'end')
    #clear address
    street.delete(0, 'end')
    town.delete(0, 'end')
    state.delete(0, 'end')
    zipcode.delete(0, 'end')
    #clear priority and class/access/availability
    access.set('Public')
    pri.set('None')
    #clear recurance feilds
    occurence.set("--")
    freq.set("----")
    interval.set("--")
    su.set(0)
    mo.set(0)
    tu.set(0)
    we.set(0)
    th.set(0)
    fr.set(0)
    sa.set(0)
    #clear responce lables
    extraFeedback['text'] = " "
    testFeedback['text'] = " "
    return
    
    
# fill the feilds based on the info in the existing event
def veiw():
    testFeedback['text'] = "to be implamented: view info of existing event"
    # this will probably be tied to a button by the input eventName feild
    # try to open file based on user input > fill in feilds read from opened file
    # exeption IOError or blank > lable "please fill out requiered fields then press the save button at the bottom"
    # lable['text'] = "to be implamented: view info of existing event"
    return


#check if create a new file or edit an existing file when button is pressed
def validate():
    lable['text'] =  " "
    testFeedback['text'] = " "
    extraFeedback['text'] = " "
    try:
        eventExisting = open(eventName.get(), 'r')
        lable['text'] = "File already exists, please rename"
    except IOError:  
        try:
            if check() == True:
                eventNew = open(eventName.get(), 'w')
                lable['text'] = "New Event %s Created"%eventName.get()
                create(eventNew)
                clearFeilds()  
        except IOError:
            lable['text'] = "No file specified"
```
